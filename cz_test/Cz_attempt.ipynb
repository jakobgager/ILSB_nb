```python
class dispeq():
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def _repr_latex_(self):
        return '${0}={1}$'.format(sp.latex(self.a),sp.latex(self.b))
```

```python
def evalsym(f, x, xs, subdict):
    fv = []
    for xv in x:
        fv.append(f.subs({xs:xv}).subs(subdict).evalf(5))
    return fv
```

## Cohesive zone model (Bouvard)

###Backward Euler
$$\dot{D}_{i+1} = 
f(D\_{i+1}, T\_{i+1})$$
requires iteration!

```python
TN1 = k*(1-Dip)*UN#/d0   # ...modified by greg
dispeq('T_{N;i+1}^j', TN1)
```

```python
MacB = sp.Piecewise((x,x>0),(0,True))
MacB
```

```python
D1 = (0.6*(1-Dip)**1.8*MacB.subs({x:(T/(1-Dip)-T0)})**1.4*(dN/d0))+Di
dispeq('D_{i+1}^{j+1}',D1)
```

```python
def iterate(TN1, D1, subdict, thres=0.5, maxiter=10, verbose=True):
    niter = 0
    diff = 2*thres
    subdict[Dip] = subdict[Di]
    while (diff > thres) and (niter < maxiter):
        Ti1 = TN1.evalf(8, subs=subdict)
        if subdict[dN] <= 0:
            Di1 = subdict[Di]
        else:
            Di1 = D1.evalf(8, subs=dict({T:Ti1}, **subdict))
        if Di1 > 1: 
            Di1=1
            Ti2=0
            diff = 0 # abort
        else:
            subdict[Dip] = Di1  # set for evaluation of Traction at end of increment
            Ti2 = TN1.evalf(8, subs=subdict)
            diff = abs((Ti1-Ti2)/Ti2)
        if verbose:
            print niter, Ti1, Ti2, Di1, diff
        niter += 1
    return Ti2, Di1, niter
```

```python
iterate(TN1, D1, {k:2.8E6, d0:1, dN:1E-6, Di:0., T0:10, UN:1.35E-4}, thres=0.0001, maxiter=10)
```

## Test Ansys Fortran implementation

Here, the traction of the previous increment is used to evaluate the current damage rate.
$$\dot{D}_{i+1} = 
f(D\_{i}, T\_{i})$$

```python
def iterate2(TN1, D1, subdict, verbose=True):
    subdict[Dip] = subdict[Di]
    Ti1 = subdict[TN0]
    if subdict[dN] <= 0:
        Di1 = subdict[Di]
    else:
        Di1 = D1.evalf(8, subs=dict({T:Ti1}, **subdict))
    if Di1>1: Di1=1
    subdict[Dip] = Di1 # to evaluate the current Traction
    Ti2 = TN1.evalf(8, subs=subdict)
    if verbose:
        print Ti1, Ti2, Di1
    return Ti2, Di1
```

```python
iterate2(TN1, D1, {k:k0, d0:1, dN:1.02E-6, Di:0., T0:10, UN:0, TN0:0})
```
