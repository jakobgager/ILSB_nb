### However for small waviness the offset produces quite good results

### To derive the correct path the normal in each point is to be used

```python
hx = x + dx
display(hx)
hy = f + dy
display(hy)
```

```python
#distance
r1 = 0.2
figure(1,(8,6))
plot(func(hx.subs({r:r1}),xv),func(hy.subs({r:r1}),xv), 'b-', label='para')
plot(xv,func(f,xv)+r1,'r-', label='off')
plot(xv,func(f,xv),'g-', label='func')
ax= gca()
ax.set_aspect('equal')
xlim(4,6)
ylim(-4,-2)
legend();
```
