## Helper functions

Class to pretty print variables (as equation) using latex.

```python
class dispeq():
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def _repr_latex_(self):
        return '${0}={1}$'.format(sp.latex(self.a),sp.latex(self.b))
```

The original numpy einsum method does not work with sympy object arrays. Therefore, another implementation is used.    
(Found: http://stackoverflow.com/questions/15606937/how-do-i-get-numpy-einsum-to-play-well-with-sympy)

```python
def einsum(string, *args):
    index_groups = map(list, string.split(','))
    assert len(index_groups) == len(args)
    tensor_indices_tuples = zip(index_groups, args)
    return reduce(einsum_for_two, tensor_indices_tuples)[1]

def einsum_for_two(tensor_indices1, tensor_indices2):
    string1, tensor1 = tensor_indices1
    string2, tensor2 = tensor_indices2
    sum_over_indices = set(string1).intersection(set(string2))
    new_string = string1 + string2
    axes = ([], [])
    for i in sum_over_indices:
        new_string.remove(i)
        new_string.remove(i)
        axes[0].append(string1.index(i))
        axes[1].append(string2.index(i))
    return new_string, np.tensordot(tensor1, tensor2, axes)
```

Method to convert a 2nd order (elasticity) tensor to is original 4th order form.

**Attention** the proposed form is only valid for 2D matrices! If a 3D representation is
required, some additional terms have to be added (see e.g. Cowin1995)

The 4th order tensor can be used like

    sp.Matrix(einsum('ijkl,kl',S,np.array(Sm)))

```python
def conv2to4order(T2, tensor='C', verbose=False):
    """Converts a 2nd order (matrix) representation of the 2D elasticity tensor (S) or the
    2D compliance tensor (S) to its original 4th order version.

    S ... compliance matrix
    C ... elasticity matrix

    requires: numpy as np
    """
    dim = 2
    if tensor=='S':
        f1 = 1
        f2 = sp.Rational(1,2)
    elif tensor=='C':
        f1 = 1
        f2 = 1
    else:
        print 'Only type S and C is allowed.'
    #T ... 4th order tensor (output)
    T4=np.empty([2,2,2,2],dtype=object)

    #mapping = [(1,1),(2,2),(1,2)]
    # mapping: index2rank, index4rank, index4rank, factor
    mapping = [(0,0,0,f1),(1,1,1,f1),(2,0,1,f2),(2,1,0,f2)]
    
    if verbose: 
        print 'mapping'
        print '\t2nd          4th'
    
    for i,n,m,f in mapping:
        for j,o,p,q in mapping:
            T4[n,m,o,p] = f*q*T2[i,j]
            if verbose:
                print '{3}\t{2}{1} -> {2}{0}'.format((n,m,o,p),(i,j),tensor,f*q)
    return T4
```

## Elasticity tensor

In the following the elastic relations of a plane stress 2D media are defined.

### Isotropic symmetry

The compliance matrix is defined as:

```python
Si = sp.Matrix([[1/Em, -nu/Em, 0],[-nu/Em, 1/Em, 0],[0, 0, 1/(Gm)]])
dispeq(Ss,Si)
```

The elasticity matrix can be computed as:

```python
Ci = Si.inv()
Ci.simplify()
dispeq(Cs,Ci)
```

To be able to relate the 2nd order stress and strain tensors the 4th order elasticity (compliance) tensor has to be
used.

```python
S4i = conv2to4order(Si,tensor='S')
C4i = conv2to4order(Ci, tensor='C')
```

### Deformation gradient $\mathcal F$

The deformation gradient can be expressed as $\mathcal F=dx\ dX^{-1}$

```python
F = dx*dX.inv()
F.simplify()
dispeq(Fs,F)
```

#### det($\mathcal{F}$) = $J$

```python
J = F.det()
dispeq(Js,J)
```

### Displacement gradient $\mathcal G$

```python
G = F - sp.eye(2)
G.simplify()
dispeq(Gs,G)
```

### linearized strain $\mathcal{E}_\mathrm{lin}$

```python
Elin = sp.S(1)/2*(G+G.transpose())
dispeq(Elins,Elin)
```

### Green-Lagrange strain $\mathsf{E}^\mathrm{GL}$

```python
EGL = sp.S(1)/2*(F.transpose()*F-sp.eye(2))
EGL.simplify()
dispeq(EGLs,EGL)
```

### right stretch $\mathsf{U}^2$

```python
U2 = F*F.transpose()
dispeq(U2s,U2)
```

### log. strain $\mathcal{E}_{\mathrm{log}}$

computes as $\mathcal{E}_{\mathrm{log}} = \log \mathsf{U}$ with
$\mathsf{U} = \sqrt{\mathcal{F}\mathcal{F}^\mathrm T}$. <BR>
The sqare root of a matrix $A$ can be computed using diagonalization (see http://en.wikipedia.org/wiki/Square_root_of_a_matrix) as
$$\sqrt{A} = VSV^{-1}\;.$$ 
Here, $V$ is the matrix of the normalized eigenvectors and $S$ is a diagonal matrix 
of the eigenvalue squareroots.

```python
U2EV = U2.eigenvects()
U2EVs = U2EV[0][2][0].normalized().row_join(U2EV[1][2][0].normalized())
#U2EV
```

```python
Sm = sp.diag(*[sp.log(sp.sqrt(U2EV[i][0])) for i in [0,1]])#.simplify()
Sm.simplify()
```

### Variation of the potential

To compute the deformed shape the principle of the stationary potential is applied. Due to the intrinsic nature
of the problem a direct solution of the equilibrium conditions is not applicable.

$$w_{\mathrm{el}}=\frac{1}{2}\int_V\sigma:\varepsilon dV$$

```python
# strain energy density
SE = sp.S(1)/2*sum(Slin.multiply_elementwise(Elin))
SE.simplify()
# strain energy
wellin = SE*V0
dispeq(wells,wellin)
```

#### External work

The external work derived at point C can be expressed as
$$w_{\mathrm{ext}} = -\vec{F^c} \cdot \vec{u^{c}}$$
with
$$\vec{F^c}=\sigma\ \vec{n^c} da\;.$$

The current area, $da$, can be computed using Nanson's relation:
$$da\ \vec n=J\ dA\ \mathcal{F}^{-T} \vec N$$
However, if a 2D deformation gradient is used this formular may be inappropriate. The assess this statement both, a 2D and a 3D, representation is applied here.

The rotation matrix is taken from the sympy package (attention: here the transposed is used)

```python
from sympy.matrices import rot_axis3
dispeq('R',rot_axis3(phi).transpose())
```

```python
N = rot_axis3(phi).transpose()[:2,:2]*sp.Matrix([1,0])
dispeq('\\vec N',N)
```

```python
# asssuming a loading in 45deg
dA = sp.sqrt(l1**2+l2**3)*l3
dan = J*dA*F.transpose().inv()*N
dispeq('da\ \\vec n',dan)
```

To test the results some numbers are introduced.

```python
dAn = dA.evalf(5,subs={l1:1,l2:1,l3:1})
dispeq('dA',dAn)
```

```python
dan1 = dan.subs({l1:1,l2:1,l3:1,ua1:0, ua2:0.2,ub1:0.2, ub2:0, phi:sp.pi/4})
dispeq('da\ \\vec n',dan1)
```

```python
da = (dan1[0]/N[0]).evalf(5, subs={phi:sp.pi/4})
dispeq('da',da)
```

From geometry considerations follows for the present case

```python
da_geo = sp.sqrt(2*(l1-ub1)**2)*l3
dispeq('da_{geo}',da_geo.evalf(5, subs={l1:1, l3:1, ub1:0.2}))
```
