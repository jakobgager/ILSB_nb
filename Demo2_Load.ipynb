# Derive a somehow reasonable pressure field for the EPD demonstrator

```python
xc1 = 15
xc2 = -484.77
```

```python
xv = linspace(xc2, 0, 200)
```

Current pressure field:

```python
def pf1(x):
    return pow((x-xc1),2)/1000000
```

```python
yv = pf1(xv)
```

```python
plot(xv, yv);
autoscale(False)
plot([xc1]*2, [0,10],'k')
plot([xc2]*2, [0,10],'k')
fill_between(xv,yv,where=xv<=xc1, facecolor='0.8', edgecolor='none')
```

Total preasure per mm circumference:

```python
quad(pf1, xc2, xc1)
```

Optimized preasure field

```python
def pf2(x):
    return 7.8/90*(np.cos((x-xc1)/((xc2-xc1)/pi))-1)

yv2 = pf2(xv)
```

```python
plot(xv, yv2);
autoscale(False)
plot([xc1]*2, [0,10],'k')
plot([xc2]*2, [0,10],'k')
fill_between(xv,yv2,where=xv<=xc1, facecolor='0.8', edgecolor='none')
```

Total preasure per mm circumference

```python
quad(pf2, xc2, xc1)
```
